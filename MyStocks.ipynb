Import the necessary libraries

```python
import pandas as pd
import numpy as np
import datetime
%matplotlib inline
import os
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline
```

Define start and end dates

```python
total_today=0
startdate='2020-11-10'
enddate='2020-12-30'
```

Define your stock ticker list that you need to follow

```python
stocktotal=pd.DataFrame()
mystocks=['AEYE','ALGM','PLUG','NUAN','FUV']
numberofstocks=[10,10,10,10,13]
color2=['red','green','blue','yellow','orange','black','turquoise','indigo','turquoise','plum']
```

Get stock data from yahoo finance
and plot the stocks


```python
i=0
for stock in mystocks:
    ticker=yf.download(stock,start=startdate,end=enddate,progress=False)
    stocktotal[stock]=ticker['Close']*numberofstocks[i]
    stocktotal[stock +' Return'] = stocktotal[stock].pct_change()
    ticker['Close'].plot(figsize=(10,5),color=color2[i],label=stock)
    print (stock)
    plt.legend()
    i=i+1
```

Total the stocks and plot the total

```python
stocktotal['Total']=stocktotal.sum(axis=1)
```

```python
stocktotal['Total'].plot(figsize=(10,5),color='purple', marker='o')
plt.xticks(rotation=15)
plt.title('My Stocks')
```
